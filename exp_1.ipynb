```python
import numpy as np
import scipy

from matplotlib import pyplot as plt
from datetime import datetime

import optimization
import oracles
from plot_trajectory_2d import plot_levels, plot_trajectory

TOOLS = ["Wolfe", "Constant", "Armijo"]

xrange = [-10, 10]
yrange = [-10, 10]
levels = np.arange(10) ** 3

def plottingQuadratic(oracle, x_0, tool, xrange=None, yrange=None, levels=None):
    plt.figure(figsize=(15, 9))
    plt.grid(alpha=0)
    [x_star, msg, history] = optimization.gradient_descent(
        oracle, x_0, line_search_options=optimization.LineSearchTool(tool), trace=True)
    plot_levels(oracle.func, xrange=xrange, yrange=yrange, levels=levels)
    plot_trajectory(oracle.func, history['x'])
    plt.xlabel("x", size = 14)
    plt.ylabel("y", size = 14)
    plt.savefig("pictures/{0}_{1}".format(oracle.A, tool))

oracle = oracles.QuadraticOracle(np.array([[1, 0], [0, 10]]), np.array([2, 4]))
x_0 = np.array([9, 1])
for tool in TOOLS:
    plottingQuadratic(oracle, x_0, tool, xrange, yrange, levels)
```
